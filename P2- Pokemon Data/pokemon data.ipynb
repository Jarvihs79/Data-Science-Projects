```python
import pandas as pd
```

```python
poke= pd.read_csv("Pokemon.csv")
```

```python
poke
```

```python
pokehp= poke.nlargest(10,["HP"])
```

```python
pokehp.plot(kind= "bar", x= "Name", y= "HP")
```

```python
pokehp
```

```python
pokehp["Name"]
```

```python
barhpoke=pokehp.plot(kind= "barh", x= "Name", y= "HP", color=("green"),title="Top 10 pokemon in terms of their HP",figsize=(8,6)).invert_yaxis()
```
